```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
```

```python
slope = 0.5
intercept = 4

y = x*slope + intercept
```

```python
y
```

```python
plt.figure()
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
# area and price

area = np.array([700,750,800,900,1000])  
price = np.array([20,30,33,45,80])
```

```python
plt.figure()
plt.scatter(area,price)
plt.show()
```

```python
# random slope and intercept
slope = 0.5
intercept = 10


predicted_price = area*slope + intercept
```

```python
plt.figure()
plt.scatter(area,price)
plt.plot(area,predicted_price)
plt.show()
```

```python
area_mean = np.mean(area)
price_mean = np.mean(price)
```

```python
slope = np.sum((area - area_mean) * (price - price_mean)) / np.sum((area - area_mean)**2)
```

```python
slope
```

```python
intercept = price_mean - (slope * area_mean)
```

```python
intercept
```

```python
predicted_price = area*slope + intercept
```

```python
plt.figure()
plt.scatter(area,price)
plt.plot(area,predicted_price)
plt.show()
```
